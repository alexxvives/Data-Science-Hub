## Autoregressive Model (AR)

An **Autoregressive model** predicts the current value of a time series using a linear combination of $t$ past values plus a random error term.  
An AR($p$) model of order $p$ is written as:

$$
y_t = c + \sum_{i=1}^{p} \phi_i\, y_{t-i} + \varepsilon_t
$$

- $y_t$: value at time $t$  
- $c$: intercept  
- $\phi_i$: coefficient of the $i$-th lag  
- $p$: number of lags used  
- $\varepsilon_t$: white noise (zero mean, constant variance)  

**Example:**  
- AR(1): $y_t = c + \phi_1 y_{t-1} + \varepsilon_t$  
- AR(2): $y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \varepsilon_t$  

Best suited for **stationary** series where recent values **influence future** ones.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
```

```python
np.random.seed(42) # For reproducibility
def simulate_ar(phi, n, c, burn_amount):
    p = len(phi)
    N = n + burn_amount # Extend n with the to-be-burnt datapoints

    y = np.zeros(N, dtype=float)
    y[:p] = np.random.normal(0.0, 1, size=p) # initialize first p values
    
    eps = np.random.normal(0.0, 1, size=N)
    
    for t in range(p, N):
        past = y[t-p:t]  # Extract the last p datapoints
        y[t] = c + np.dot(phi, past[::-1]) + eps[t]
        
    return y[burn_amount:] # Return values after burn

# Parameters for AR(2)
phi = np.array([0.5, -0.3]) # Coefficients
c = 0.3 # Intercept
n = 200 # Datapoints
burn_amount = 150 # Initial simulated points to throw away (allows simulated series to settle into its natural dynamics)

# Create synthetic data
y = simulate_ar(phi=phi, n=n, c=c, burn_amount=burn_amount)
df = pd.DataFrame({"t": np.arange(1, n+1), "data": y}).set_index("t")

# Plot
plt.figure(figsize=(12, 4))
plt.plot(df.index, df["data"])
plt.title(f"Synthetic AR(2) Time Series: y_t = {c} + {phi[0]}*y_(t-1) + {phi[1]}*y_(t-2) + eps_t")
plt.xlabel("Time (t)")
plt.ylabel("y_t")
plt.tight_layout()
plt.show()
```

#### Stationarity Assumption
The stationarity assumption is valid because the ADF test rejects the null hypothesis of non-stationarity, indicating the data is stationary.

```python
from statsmodels.tsa.stattools import adfuller
p_value = adfuller(df['data'], autolag='AIC', regression='c')[1]
if p_value < 0.05: print('Data is stationary') 
else: print('Data is not stationary')
```

### Manual Calculation of Autoregressive Model (AR)

```python
def forecast_recursive(y, c, phi, h, p):
    last_values = list(y[-p:])
    out = []
    for _ in range(h):
        y_next = c + np.dot(phi, last_values[-p:][::-1])
        last_values.append(y_next)
        out.append(y_next)
    return np.array(out)

p = 2 # AR order
h = 20 # steps to forecast beyond the sample
train_ratio = 0.8 # train/test split
train_size = int(train_ratio * len(Y))
n = len(df["data"])

# Build design
X, Y = pd.DataFrame({'const': 1.0}, index=df["data"].index[p:]), df["data"].iloc[p:]
for lag in range(1, p + 1):
    X[f'lag_t-{lag}'] = df["data"].shift(lag).iloc[p:].values

X_train, Y_train = X[:train_size], Y[:train_size]
X_test, Y_test = X[train_size:], Y[train_size:]

# Solve for coefficients using an SVD-based least-squares solver
beta = np.linalg.lstsq(X_train, Y_train, rcond=None)[0]
c_hat, phi_hat = beta[0], beta[1:]

# Build predictions for training and testing data
Y_train_pred = X_train @ beta
Y_test_pred = X_test @ beta

# Run AR(2) forecast
y_forecast = forecast_recursive(df["data"], c_hat, phi_hat, h, p)

plt.figure(figsize=(12, 5))

# Plots
plt.plot(np.arange(n), y, label="Observed", linewidth=1.5)

t_values_train = np.arange(p, n)[:train_size]
plt.plot(t_values_train, Y_train_pred, label="Fitted (train)", linestyle="--")

t_values_test = np.arange(p, n)[train_size:]
plt.plot(t_values_test, Y_test_pred, label="1-step ahead (test)", linestyle="--")

t_values_forecast = np.arange(n, n + h)
plt.plot(t_values_forecast, y_forecast, label=f"{h}-step forecast (recursive)", linewidth=1.8)

# Create Strain/test split marker
split_t = train_size + 1
plt.axvline(split_t - 0.5, color="gray", alpha=0.4)
plt.text(split_t, plt.ylim()[1], " train | test ", va="top", ha="left", color="gray")

plt.title(f"Autoregressive Model - AR({p})")
plt.xlabel("Time index")
plt.ylabel("Value")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
