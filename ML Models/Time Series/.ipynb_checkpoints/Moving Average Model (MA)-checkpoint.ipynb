## Moving Average Model (MA)

A **Moving Average model** predicts the current value of a time series as the mean plus a linear combination of the **last \(q\) shocks (errors)**.  <br>
An MA($q$) model of order $q$ is written as:

$$
y_t = \mu + \varepsilon_t + \sum_{i=1}^{q} \theta_i\, \varepsilon_{t-i}
$$

- $y_t$: value at time \(t\)  
- $\mu$: mean level (intercept)  
- $\theta_i$: coefficient of the \(i\)-th past error  
- $q$: number of past errors used  
- $\varepsilon_t$: white noise (zero mean, constant variance)

**Example:**  
- MA(1): $y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1}$  
- MA(2): $y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2}$

Best suited for **stationary** series with **short-lived, shock-driven patterns**.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats
from statsmodels.tsa.arima.model import ARIMA

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
```

```python
np.random.seed(42)  # For reproducibility
def simulate_ma(theta, n, mu, burn_amount):
    q = len(theta)
    N = n + burn_amount # Extend n with the to-be-burnt datapoints

    y = np.zeros(N, dtype=float)
    eps = np.random.normal(0.0, 1, size=N)

    for t in range(N):
        base = mu + eps[t]
        k = min(q, t)
        past_eps = eps[t - np.arange(1, k+1)]  # newest to oldest
        base += np.dot(theta[:k], past_eps)
        y[t] = base

    return y[burn_amount:] # Return values after burn

# Parameters for MA(2)
theta = np.array([0.6, -0.3]) # Coefficients
mu = 0.3 # Mean (intercept)
n = 200 # Datapoints
burn_amount = 150 # Initial simulated points to throw away (allows simulated series to settle into its natural dynamics)

# Create synthetic data
y = simulate_ma(theta=theta, n=n, mu=mu, burn_amount=burn_amount)
df = pd.DataFrame({"t": np.arange(1, n+1), "data": y}).set_index("t")

# Plot
plt.figure(figsize=(12, 4))
plt.plot(df.index, df["data"])
plt.title(f"Synthetic MA(2) Time Series: y_t = {mu} + eps_t + {theta[0]}*eps_(t-1) + {theta[1]}*eps_(t-2)")
plt.xlabel("Time (t)")
plt.ylabel("y_t")
plt.tight_layout()
plt.show()
```

#### Stationarity Assumption
The stationarity assumption is valid because the ADF test rejects the null hypothesis of non-stationarity, indicating the data is stationary.

```python
from statsmodels.tsa.stattools import adfuller
p_value = adfuller(df['data'], autolag='AIC', regression='c')[1]
if p_value < 0.05: print('Data is stationary') 
else: print('Data is not stationary')
```

### Manual Calculation of Moving Average (MA)

```python
def ma_future_forecast_dot(mu, theta, eps_hat, h):

    theta = np.asarray(theta, float)
    q = len(theta)
    eps_hat = np.asarray(eps_hat, float)

    out = np.empty(h, float)
    for k in range(1, h+1):
        m = max(q - (k - 1), 0) # number of residuals still contributing
        recent = eps_hat[-m:][::-1] if m > 0 else np.empty(0)
        out[k-1] = mu + (np.dot(theta[:m], recent) if m > 0 else 0.0)
    return out

q = 2 # MA order
h = 20 # steps to forecast beyond the sample
train_ratio = 0.8 # train/test split
train_size = int(train_ratio * n)

# Fit MA(2) on the training data
MA_model = ARIMA(df["data"].iloc[:train_size], order=(0, 0, q), trend='c').fit()

# Apply fitted model to the full data (training + testing)
applied =  MA_model.apply(df["data"], refit=False)

# Get training/testing predictions + epsilons (used for MA forecast)
y_train_fit = applied.fittedvalues.iloc[:train_size] # Get training predictions
y_test_pred = applied.fittedvalues.iloc[train_size:] # Get testing predictions
eps_hat  = applied.resid   

# Run MA(2) forecast
mu_hat = MA_model.params.get('const', MA_model.params.get('intercept', 0.0))
theta_hat = MA_model.maparams 
y_forecast = ma_future_forecast_dot(mu_hat, theta_hat, eps_hat.values, h)

# Plots
plt.figure(figsize=(12, 5))
plt.plot(np.arange(n), df["data"], label="Observed", linewidth=1.5)
plt.plot(np.arange(train_size),  y_train_fit.values, linestyle="--", label="Fitted (train)")
plt.plot(np.arange(train_size, n),  y_test_pred.values, linestyle="--", label="1-step ahead (test)")
plt.plot(np.arange(n, n + h), y_forecast, linewidth=1.8, label=f"{h}-step forecast")

# Train/Test split marker
plt.axvline(train_size - 0.5, color="gray", alpha=0.4)
plt.text(train_size, plt.ylim()[1], " train | test ", va="top", ha="left", color="gray")

plt.title(f"Moving Average Model  MA({q})")
plt.xlabel("Time index")
plt.ylabel("Value")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
