## SARIMA

#### Description

A **SARIMA** (Seasonal AutoRegressive Integrated Moving Average) model extends **ARIMA** by explicitly modeling **seasonal patterns** in addition to short‑term dynamics and trends.

**What SARIMA does (plain language):**
- It **stabilizes** the series by differencing away trend and repeating seasonal drift (the **I** part).
- It captures **persistence from past values** (the **AR** part).
- It captures **echoes of past surprises** (the **MA** part).
- It repeats those AR/MA ideas specifically at the **seasonal period $s$** to model patterns that recur every $s$ steps (the **seasonal** part).

**Notation:**
$$
\text{SARIMA}(p,d,q)\,(P,D,Q)_s
$$
- $(p,d,q)$ → nonseasonal AR, differencing, MA orders  
- $(P,D,Q)_s$ → seasonal AR, differencing, MA orders with season length $s$ (e.g., $s=12$ for monthly data with yearly seasonality)

---

### **AR — AutoRegressive (order $p$)**

The autoregressive (AR) component models the current value of the series as a weighted combination of the last $p$ past values. In other words, today’s value is explained, at least in part, by what happened in the recent past. *If the series has been high for the last few time steps, it is more likely to remain high.*

**Mathematical form:**
$$
y_t = c + \sum_{i=1}^{p} \phi_i \, y_{t-i} + \varepsilon_t
$$
where $\phi_i$ are the autoregressive coefficients, $c$ is a constant term, and $\varepsilon_t$ is the random error at time $t$.

**How to recognize it:** In an AR($p$) process, the **PACF:** shows clear, significant spikes up to lag $p$ and then drops near zero for higher lags. The highest lag with a significant spike in the PACF is typically taken as the AR order $p$.

---

### **I — Integrated (nonseasonal differencing order $d$ and seasonal differencing order $D$)**

The integrated (I) component specifies how many times the series must be differenced to remove patterns that prevent stationarity. There are two types:

- **Nonseasonal differencing ($d$)** removes long-term trends by subtracting each value from the previous one:
$$
(1 - B)^d \, y_t
$$
It targets gradual drift in the mean so the series no longer trends upward or downward over time.

- **Seasonal differencing ($D$)** removes repeating seasonal patterns by subtracting each value from the value one full season earlier:
$$
(1 - B^s)^D \, y_t
$$
where $s$ is the length of the seasonal cycle (e.g., $s=12$ for monthly data with yearly seasonality).

---

### **MA — Moving Average (order $q$)**

The moving average (MA) component models the current value of the series as a weighted combination of the last $q$ past forecast errors (shocks). In other words, today’s value is influenced by recent unexpected changes — the lingering effects of surprises in the data.

**Mathematical form:**
$$
y_t = \mu + \varepsilon_t + \sum_{j=1}^{q} \theta_j \, \varepsilon_{t-j}
$$
where $\theta_j$ are the moving average coefficients, $\mu$ is the mean of the series, and $\varepsilon_t$ is the random error at time $t$.

**How to recognize it:** In an MA($q$) process, the **ACF** shows clear, significant spikes up to lag $q$ and then drops near zero for higher lags. The highest lag with a significant spike in the ACF is typically taken as the MA order $q$.


---

### **Seasonal part — $(P,D,Q)_s$**

**Idea:** Apply AR, I, and MA at **seasonal lags** $s, 2s, \dots$ to capture repeating patterns.

- **Seasonal AR ($P$):** includes $y_{t-s}, y_{t-2s}, \dots$
- **Seasonal differencing ($D$):** $(1 - B^s)^D$ removes repeating seasonal pattern
- **Seasonal MA ($Q$):** includes $\varepsilon_{t-s}, \varepsilon_{t-2s}, \dots$

**Example:** With monthly data and yearly seasonality ($s=12$), a seasonal AR(1) means **this January** depends on **last January**.

---

### Full equation

$$
y'_t = c + \sum_{n=1}^p \phi_n y'_{t-n} 
      + \sum_{n=1}^q \theta_n \varepsilon_{t-n} 
      + \sum_{n=1}^P \eta_n y'_{t - mn} 
      + \sum_{n=1}^Q \omega_n \varepsilon_{t - mn} 
      + \varepsilon_t
$$

Where:
- y’: differenced time series, through both regular, d, and seasonal, D, differencing
- P: number of seasonal auto-regressors
- ω: coefficients of the seasonal autoregressive components
- Q: number of seasonal moving-average components
- η: coefficients of the seasonal forecast errors
- m: length of season

#### SARIMA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import boxcox
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from scipy.special import inv_boxcox
from statsmodels.tsa.arima.model import ARIMA
import itertools
from statsmodels.tsa.stattools import adfuller

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
```

```python
passengers = [
    112,118,132,129,121,135,148,148,136,119,104,118,115,126,141,135,125,149,170,170,158,133,114,140,145,150,178,163,172,178,199,199,184,162,146,166,
    171,180,193,181,183,218,230,242,209,191,172,194, 196,196,236,235,229,243,264,272,237,211,180,201,204,188,235,227,234,264,302,293,259,229,203,229,
    242,233,267,269,270,315,364,347,312,274,237,278,284,277,317,313,318,374,413,405,355,306,271,306,315,301,356,348,355,422,465,467,404,347,305,336,
    340,318,362,348,363,435,491,505,404,359,310,337,360,342,406,396,420,472,548,559,463,407,362,405,417,391,419,461,472,535,622,606,508,461,390,432]
dates = pd.date_range(start="1949-01", periods=len(passengers), freq="ME")

# Create DataFrame
df = pd.DataFrame({"Month": dates, "Passengers": passengers})
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

plt.figure(figsize=(12, 3))
sns.lineplot(data=df, x="Month", y="Passengers")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Passengers", fontsize=12)
plt.title("Airline Passengers", fontsize=16)
plt.show()
```

```python
result = adfuller(df['Passengers'])
if result[1] > 0.05: print('Data is not stationary')
else: print('Data is stationary')
```

SARIMA models need to have **stationary data** to model and forecast the time series.<br> *A stationarity time series does not exhibit any long-term trend or clear seasonality, and its statistical properties, such as mean and variance, remain constant over time.*

To produce a stationary time series we need to **stabilize the mean and variance**. <br>
The **mean** can be stabilized through **differencing** and the number of differentiations applied is d (or D in the case of seasonal differencing). <br>
The **variance** can be stabilized through transformations such as the logarithmic or Box-Cox transform, this makes the seasonal fluctuations occur on a similar level every season.

<img src="../Images/SARIMA.png" alt="My Image Caption" width="1000" height="1000">

Our data has both its mean and variance increasing overtime, therefore we need to **apply a transformation** (such as Box-Cox) to stabilize the variance and then **differentiate** (to stabilize the mean).

```python
# Make the data stationary
df['Passengers_Boxcox'], lamba_ = boxcox(df['Passengers']) # Stabilize variance
df["Passenger_diff"] = df["Passengers_Boxcox"].diff() # Stabilize mean
df = df.dropna()

plt.figure(figsize=(12, 3))
sns.lineplot(data=df, x="Month", y="Passenger_diff")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Passengers", fontsize=12)
plt.title("Airline Passengers", fontsize=16)
plt.show()
```

Once stationary, we then need to **deduce the best orders**, (p, d, q) and (P, D, Q)m, for our model.<br> 
The simplest ones to calculate are $d$ (non-seasonal differencing) and $D$ (seasonal differencing). <br>
This can be deduced through the *Augmented Dickey-Fuller (ADF)* statistical test that deduces whether a time series is stationary or not.

The autoregressive and moving-average orders (p, q, P, Q) can be computed by analyzing the **Partial Autocorrelation Function (PACF)** and **Autocorrelation Function (ACF)** respectively. The idea behind this technique is to plot a correlogram of the autoregressors and moving-average value and deduce which ones are statistically significant. The significant ones indicate that they have a substantial impact on the forecast.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
plot_acf(df['Passenger_diff'], ax=axes[0], lags=48)
plot_pacf(df['Passenger_diff'], ax=axes[1], lags=48)
sm.tsa.seasonal_decompose(df['Passenger_diff'], model='additive').plot();
plt.show()
```

Looking at the ACF, there is a clear seasonality at month 12

```python
df["Passenger_diff_seasonality"] = df["Passenger_diff"].diff(12)
df = df.dropna()

plt.figure(figsize=(12, 3))
sns.lineplot(data=df, x="Month", y="Passenger_diff_seasonality")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Passengers", fontsize=12)
plt.title("Airline Passengers", fontsize=16)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
plot_acf(df['Passenger_diff_seasonality'], ax=axes[0], lags=48)
plot_pacf(df['Passenger_diff_seasonality'], ax=axes[1], lags=48)
sm.tsa.seasonal_decompose(df['Passenger_diff_seasonality'], model='additive').plot();
plt.show()
```

The **Partial Autocorrelation Function (PACF)** measures the **direct correlation** between a time series and its own lagged values, after removing the effects of any shorter lags. 

The **Autocorrelation Function (ACF)** measures the correlation between a time series and its own past values, **without removing the influence of intermediate lags**.

### Model order selection from ACF/PACF

**Non-seasonal orders** $(p, d, q)$:

- $(d = 1)$: a first non-seasonal difference was applied before seasonal differencing, removing the trend.
- **AR order $(p)$**:  
  The PACF has a significant spike at lag 1, then mostly inside the confidence bands → $p = 1$.
- **MA order $(q)$**:  
  The ACF shows damped exponential/oscillation → $q = 0$.

**Seasonal orders** $(P, D, Q, m)$:

- \(D = 1, m = 12\): a seasonal difference with 12 months period was applied to remove seasonality.
- **Seasonal AR order \(P\)**:  
  PACF shows a spike at lag $m$ → $P = 1$.
- **Seasonal MA order \(Q\)**:  
  ACF shows a spike at lag $m$ → $Q = 1$.



**Initial candidate model:**
$$
\text{SARIMA}(p,d,q) \times (P,D,Q)_{m}
= (1, 1, 0) \times (1, 1, 1)_{12}
$$
- $ d = 1 $: remove trend  
- $ D = 1, \ m = 12 $: yearly seasonality  
- $ p = 1, \ q = 1 $: short-run dynamics  
- $ P = 1, \ Q = 1 $: seasonal dynamics

```python
training_size = int(len(df) * 0.7)
validation_size = int(len(df) * 0.85)
y_train = df['Passengers_Boxcox'].iloc[:training_size]
y_val = df['Passengers'].iloc[training_size:validation_size]
y_test = df['Passengers'].iloc[validation_size:]
h = len(y_val)

# p and q range from 0 to 1
pq_combos = list(itertools.product(range(4), repeat=2))[1:]
for p, q in pq_combos:
    model = ARIMA(y_train, order=(p,1,q), seasonal_order=(1,1,1,12)).fit()
    boxcox_forecasts = model.get_forecast(steps=h).predicted_mean
    forecasts = inv_boxcox(boxcox_forecasts, lamba_)
    
    RMSE = np.sqrt(np.mean((y_val.values - forecasts.values)**2))
    print(f"")
    print(f"SARIMA({p},1,{q})x(1,1,1)[12]:  AIC=%.1f  RMSE=%.2f" % (model.aic, RMSE))
```

```python
# Combine train + validation for final fit
y_trainval = df['Passengers_Boxcox'].iloc[:validation_size]
y_test = df['Passengers'].iloc[validation_size:]  # original scale for evaluation
h = len(y_test)

# Fit the SARIMA model (example: best params from validation)
model = ARIMA(y_trainval, order=(2, 1, 2), seasonal_order=(1, 1, 1, 12)).fit()

# Get forecast in Box-Cox space (+ mean and CI on boxcox forecast)
fc_res = model.get_forecast(steps=h)
prediction_mean_bc = fc_res.predicted_mean
prediction_CI_bc = fc_res.conf_int()

# Inverse Box-Cox to get back to passenger counts
fc_mean = inv_boxcox(prediction_mean_bc, lam)
CI_low = inv_boxcox(prediction_CI_bc.iloc[:, 0], lam)
CI_high = inv_boxcox(prediction_CI_bc.iloc[:, 1], lam)

# Plot
plt.figure(figsize=(10, 3))
sns.lineplot(x=y_trainval.index, y=inv_boxcox(y_trainval, lam), label='Train+Valid')
sns.lineplot(x=y_test.index, y=y_test, label='Test')
sns.lineplot(x=y_test.index, y=fc_mean, label='Forecast')
plt.fill_between(y_test.index, CI_low, CI_high, alpha=0.15, label='95% CI')
plt.xlabel("Date")
plt.ylabel("Passengers")
plt.legend()
plt.tight_layout()
plt.show()

rmse = np.sqrt(np.mean((y_test.values - fc_mean.values) ** 2))
mae = np.mean(np.abs(y_test.values - fc_mean.values))
print(f"RMSE: {rmse:.2f}   MAE: {mae:.2f}   AIC: {model.aic:.1f}   BIC: {model.bic:.1f}")
```
