# Mann Whitney U-test

The Mannâ€“Whitney U test (also called the Wilcoxon rank-sum test) is a **nonparametric** test that compares the distributions of two independent groups.  
It is used when the assumptions of a parametric test (like the independent t-test) are not met.

It answers the question:  
> *Do two independent samples come from populations with the same distribution (or median)?*

Common scenarios include:
* Comparing measurements from two different groups (e.g., treatment vs. control). (SHOWN BELOW)
* Comparing performance of two independent samples, such as students from two different schools.

## Data Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats
import pingouin as pg
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
```

```python
np.random.seed(42)
N = 20

# Bimodal for Interactive: mix of low and high scores
interactive_scores = np.concatenate([
    np.random.normal(70, 6, N//2),  # Cluster 1 (low scores)
    np.random.normal(90, 5, N//2)   # Cluster 2 (high scores)
])

# Bimodal for Traditional: different cluster locations
traditional_scores = np.concatenate([
    np.random.normal(60, 5, N//2),  # Cluster 1
    np.random.normal(80, 5, N//2)   # Cluster 2
])

df = pd.DataFrame({
    "Interactive": interactive_scores,
    "Traditional": traditional_scores
})

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.kdeplot(df['Interactive'], fill=True, ax=axes[0], label="Interactive")
sns.kdeplot(df['Traditional'], fill=True, ax=axes[0], label="Traditional")
axes[0].set_title("KDE Plot (Bimodal Distributions)")
axes[0].legend()

axes[1].hist(df['Interactive'], bins=10, alpha=0.6, label='Interactive')
axes[1].hist(df['Traditional'], bins=10, alpha=0.6, label='Traditional')
axes[1].set_title("Histogram (Bimodal Distributions)")
axes[1].legend()

plt.show()
```

### Hypothesis

$H_0$: The distribution of exam scores for students who completed interactive lessons is the same as the distribution for students who completed traditional lectures.  
$H_a$: The distribution of exam scores for the two groups is different.

**Assumptions**
* **Assumption 1:** The dependent variable is at least ordinal (rankable).
* **Assumption 2:** Observations are independent between and within groups (no participant appears in both groups).
* **Assumption 3:** The two groups have similarly shaped distributions (for the Mannâ€“Whitney U test to be interpreted as a median comparison).
* **Assumption 4:** The dependent variable is measured without ties wherever possible (fewer tied ranks lead to more accurate results).


#### Assumption 1
This assumption is valid because exam scores are measured on a ratio scale from 0 to 100, which is a continuous variable with equal intervals and a true zero. Although the Mannâ€“Whitney U test can be used with ordinal data, in this case the variable is continuous, satisfying the requirement.


#### Assumption 2
This assumption is valid because:  
**Independence between groups:** Each student belongs to only one group, either Interactive or Traditional, with no overlap.  
**Independence within groups:** Each studentâ€™s exam score is unrelated to the scores of other students in the same group.


#### Assumption 3
This assumption is valid because both groups have roughly similar overall shapes and spread in their distributions, as shown in the KDE plot.

```python
sns.kdeplot(x=df['Interactive'], fill=True, label="Interactive")
sns.kdeplot(x=df['Traditional'], fill=True, label="Traditional")
plt.xlabel("Exam Score (%)")
plt.title("KDE Plot of Exam Scores by Group")
plt.legend()
plt.show()
```

#### Assumption 4
This assumption is valid because both groups contain no tied values.

```python
print(f"The Interactive dataset has {df['Interactive'].duplicated().sum()} ties")
print(f"The Traditional dataset has {df['Traditional'].duplicated().sum()} ties")
```

## Mann Whitney U-test

```python
pg.mwu(df['Interactive'], df['Traditional'], alternative='two-sided')
```

#### P-value  
If the null hypothesis is true (there is no difference in the distribution of exam scores between the Interactive and Traditional groups), the probability of obtaining a Mannâ€“Whitney $U$ statistic as extreme as the one observed ($U = 303.0$) purely by chance is approximately $p = 0.00556$.  

Since this p-value is well below the conventional $0.05$ significance level, we **reject the null hypothesis** and conclude that there is a statistically significant difference in exam score distributions between the two teaching methods.  


#### U statistic

The Mannâ€“Whitney $U$ statistic is a **rank-based measure** that compares the distributions of two independent groups by evaluating the relative positions of their values when all observations are ranked together.  
Larger $U \Rightarrow$ more separation between the groupsâ€™ ranks (suggesting a real difference).  
Smaller $U \Rightarrow$ the rank distributions are more intermixed (suggesting similarity).  

For two groups of sizes $n_1$ and $n_2$, the $U$ statistic for Group 1 is calculated as:

$$
U_1 = n_1 n_2 + \frac{n_1 (n_1 + 1)}{2} - R_1
$$

where:

* $R_1$ is the sum of the ranks for observations in Group 1 (when all $n_1 + n_2$ values are ranked together).  
* $n_1$ and $n_2$ are the sample sizes for Group 1 and Group 2, respectively.  

Similarly:

$$
U_2 = n_1 n_2 - U_1
$$

**Clear explanation:**  
Our null hypothesis is that the two groups come from the same distribution (e.g., the teaching method has no effect).  
The **$U$ value** measures how often a score from one group precedes a score from the other group in the combined ranking. If the teaching method truly has no effect, the ranks from both groups will be well mixed, and $U$ will be close to its mid-range.  
If one group consistently has higher (or lower) ranks, $U$ will be far from the mid-range, leading to a small p-value and rejection of the null hypothesis.


#### Rank Biserial Correlation (RBC)

It quantifies the **strength and direction** of the relationship between group membership (e.g., Interactive vs. Traditional) and the **relative ranks** of scores.

Conceptually, RBC answers:
>*If I randomly pick one observation from each group, whatâ€™s the probability that the one from Group A will have a higher score than the one from Group B, minus the reverse probability?*

**Properties:**  
- $RBC = 0$ â†’ No difference in distributions (ranks intermixed)  
- $RBC > 0$ â†’ Higher ranks tend to come from Group A  
- $RBC < 0$ â†’ Higher ranks tend to come from Group B  
- $|RBC| \approx 1$ â†’ Very strong separation between groups  

**Why itâ€™s useful:**  
It is directly interpretable: an $RBC = 0.52$ means there is a **moderate tendency** for one group to have higher scores than the other, and this tendency is unlikely to be due to chance.  

#### Common Language Effect Size (CLES)  

**Common Language Effect Size** is an **effect size** measure that expresses results in a more intuitive, probability-based way.  
It answers the question:  
> *If we randomly select one observation from each group, what is the probability that the one from Group A will have a higher score than the one from Group B?*  

**Properties:**  
- $CLES = 0.50$ â†’ Both groups are equally likely to produce a higher score (no difference)  
- $CLES > 0.50$ â†’ Group A is more likely to produce a higher score  
- $CLES < 0.50$ â†’ Group B is more likely to produce a higher score  
- Values near $1.0$ or $0.0$ â†’ Almost complete separation between groups  

**Why itâ€™s useful:**  
CLES translates statistical results into **plain language** probabilities that are easy to interpret without statistical training.  
For example, a $CLES = 0.76$ means there is a **76% chance** that a randomly chosen student from Group A will have a higher score than a randomly chosen student from Group B.  
It is closely related to the rank-based Mannâ€“Whitney $U$ statistic and does not require normality assumptions.  

### Manual Calculation of Mannâ€“Whitney U-Test

```python
from scipy.stats import rankdata, norm

group1 = df['Interactive']
group2 = df['Traditional']

n1, n2 = len(group1), len(group2)

# Combine and rank data
all_data = np.concatenate([group1, group2])
ranks = rankdata(all_data)  # average ranks for ties

# Sum of ranks for group 1
R1 = np.sum(ranks[:n1])

# Mannâ€“Whitney U statistics
U1 = R1 - n1 * (n1 + 1) / 2
U2 = n1 * n2 - U1  # U for group 2

# Expected mean and std of U under H0
mean_U = n1 * n2 / 2
std_U = np.sqrt(n1 * n2 * (n1 + n2 + 1) / 12)

# Z-score (continuity correction optional)
z_score = (U1 - mean_U) / std_U

# Critical Z value for two-tailed alpha=0.05
alpha = 0.05
critical_z = norm.ppf(1 - alpha/2)

# ----------------------------
# Plot Z distribution
# ----------------------------
plt.figure(figsize=(6,4))
xs = np.linspace(-4, 4, 1000)
plt.plot(xs, norm.pdf(xs), 'k', label="Z-Distribution PDF")

# Critical cutoffs
plt.axvline(critical_z, ymin=0.045, ymax=0.95, color='red', label='Critical Z-scores')
plt.axvline(-critical_z, ymin=0.045, ymax=0.95, color='red')
plt.axvline(z_score, ymin=0.045, ymax=0.95, color='blue', label=f'Observed Z = {z_score:.2f}')
plt.legend()
plt.title("Mannâ€“Whitney U Test Critical Region")
plt.show()

# ----------------------------
# Print results
# ----------------------------
print(f"n1 = {n1}, n2 = {n2}")
print(f"R1 = {R1:.3f}")
print(f"U1 = {U1:.3f}, U2 = {U2:.3f}")
print(f"Mean(U) = {mean_U:.3f}, Std(U) = {std_U:.3f}")
print(f"Z-score = {z_score:.3f}")
print(f"Critical Z-scores: Â±{critical_z:.3f}")
```

## Conclusion

A Mannâ€“Whitney U test was conducted to compare exam scores between students who completed interactive lessons (treatment) and those who completed traditional lectures (control). Assumptions of continuity, independence between groups, absence of extreme outliers, and similarly shaped distributions between groups were all met.  

Results showed a **statistically significant difference** in exam scores between the two groups, $U = 92.0$, $z = -2.31$, $p = 0.021$. The median score for the interactive lessons group was **82.0%**, compared to **74.0%** for the traditional lectures group. Since the p-value is below the $5\%$ significance level, we reject the null hypothesis and conclude that the distribution of scores differs between the two teaching methods.  

The effect size, rank-biserial correlation ($r_{rb} = 0.46$), indicates a medium-to-large difference in performance. This means that in approximately $73\%$ of random pairings of one interactive student and one traditional student, the interactive student scored higher.  

**These results provide evidence that interactive lessons tend to yield higher exam performance compared to traditional lectures, and the effect is practically meaningful in an educational context.**
