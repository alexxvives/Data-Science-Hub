# Wilcoxon Signed-Rank W-test

The Wilcoxon signed-rank test is a **nonparametric** test that compares the distributions of two **related or paired samples**.  
It is used when the assumptions of a parametric dependent t-test or one sample t-test are not met.

It answers the question:  
> *Do two related samples (or repeated measurements on the same subjects) come from populations with the same distribution (or median)?*

Common scenarios include:
* Comparing before-and-after measurements from the same participants (e.g., pre-test vs. post-test).
* Evaluating changes in performance for individuals after a training program.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
from scipy.stats import skew

# --------------------------
# Simulated paired weight data (symmetric differences)
# --------------------------
N = 40
np.random.seed(42)

# Skewed "Before" weights
before_data = (np.random.chisquare(df=3, size=N) * 10 + 150).astype(int)

# Generate symmetric differences:
# 1) Positive magnitudes from skewed dist
# 2) Assign random  signs
magnitudes = (np.random.chisquare(df=2, size=N) * 3 + 5)
signs = np.random.choice([-1, 1], size=N)
diffs = magnitudes * signs  # Now ~50% positive, 50% negative

# Calculate "After" values
after_data = (before_data - diffs).round().astype(int)

# Create DataFrame
df = pd.DataFrame({
    "Pair": range(1, N+1),
    "Before": before_data,
    "After": after_data
}).set_index("Pair")

# Overview of dataset
print(f"The dataset contains two weight measurements (before vs. after the diet) from {df.shape[0]} participants.")
display(HTML(
    f"""
    <div style="display: flex; gap: 50px;">
        <div>{df.head().to_html()}</div>
        <div>{df.describe().T.to_html()}</div>
    </div>
    """
))

# --------------------------
# Visualizations
# --------------------------
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# KDE plot for before and after
sns.kdeplot(ax=axes[0], x=df['Before'], fill=True, label='Before', alpha=0.6)
sns.kdeplot(ax=axes[0], x=df['After'], fill=True, label='After', alpha=0.6)
axes[0].set_title("KDE Plot of Weights Before and After Diet")
axes[0].set_xlabel("Weight (lbs)")
axes[0].legend()

# Boxplot
sns.boxplot(ax=axes[1], data=df[['Before','After']], palette="pastel",
            flierprops={"markerfacecolor": "red"})
axes[1].set_title("Boxplot of Weights Before and After Diet")
axes[1].set_ylabel("Weight (lbs)")

plt.tight_layout()
plt.show()
```

### Hypothesis

$H_0$: The distribution of weight before and after completing the diet program is the same (i.e., the median difference is zero).  
$H_a$: The distribution of weight before and after completing the diet program is different (i.e., the median difference is not zero).

**Assumptions**
* **Assumption 1:** The dependent variable is continuous or ordinal.
* **Assumption 2:** Observations are paired/matched and the pairs are independent of each other.
* **Assumption 3:** The distribution of the differences between paired observations is symmetric.
* **Assumption 4:** The measurement scale is at least ordinal (ranking of values is meaningful).

*Note:* The Wilcoxon signed-rank test does **not** require normality of the differences, making it a suitable alternative to the paired t-test when this assumption is violated.


```python
# Calculate the difference in weight before and after
diff = df['After'] - df['Before']
```

#### Assumption 1
This assumption is valid because weight is measured on a ratio scale in pounds, which is a continuous variable with equal intervals and a true zero.

#### Assumption 2
This assumption is valid because:<br>
**Matching pairs**: Each Before and After measurement belongs to the same individual, so they form a natural dapair.<br>
**Indepenent Pairs**: Each participant's results are unrelated to any other participants results  no repeated subjects, no hared environments that would artificially link pairs.

#### Assumption 3

## Wilcoxon W-test

```python
pg.wilcoxon(df["Before"], df["After"], alternative="two-sided")
```

#### P-value  
If the null hypothesis is true (there is no difference in the median paired differences between the Before and After measurements), the probability of obtaining a Wilcoxon signed-rank statistic as extreme as the one observed ($W = 346.0$) purely by chance is approximately $p = 0.3974$.  

Since this p-value is well above the conventional $0.05$ significance level, we **fail to reject the null hypothesis** and conclude that there is **no statistically significant difference** in the paired measurements before and after the intervention.


The Wilcoxon signed-rank test is a **rank-based, non-parametric test** used to compare two related samples, matched samples, or repeated measurements on a single group to assess whether their population mean ranks differ.

Instead of comparing raw values, it:

1. Computes the **paired differences** between the two conditions.
2. Ignores pairs with a difference of exactly zero.
3. Ranks the absolute differences from smallest to largest, assigning average ranks to ties.
4. Sums the ranks for the positive and negative differences separately.

The **test statistic** $W$ is the smaller of these two sums of ranks.

For $n$ non-zero differences, if the null hypothesis is true (median difference = 0), the positive and negative ranks should be roughly equal in size, and $W$ will be close to its expected mid-range.

Formally:

$$
W = \min\left(\sum \text{ranks of positive differences}, \ \sum \text{ranks of negative differences}\right)
$$

**Clear explanation:**  
Our null hypothesis is that the median paired difference between the two conditions is zero (no systematic change).  
The **$W$ value** measures how much the ranks of the positive and negative differences diverge.  
If the differences are balanced around zero, $W$ will be near its mid-point, giving a large p-value and failing to reject the null.  
If one sign dominates (most differences positive or most negative), $W$ will be far from the mid-point, leading to a small p-value and rejection of the null.


#### Rank Biserial Correlation (RBC)

The **Rank-Biserial Correlation (RBC)** in the Wilcoxon signed-rank test quantifies the **strength and direction** of the relationship between the *sign* of the paired differences and their **relative ranks**.

Conceptually, RBC answers:
>*If I randomly select one non-zero paired difference from the sample, whats the probability that it is positive minus the probability that it is negative?*

**Properties:**  
- $RBC = 0$  No tendency for differences to be predominantly positive or negative (balanced signs)  
- $RBC > 0$  Positive differences tend to have larger ranks (e.g., measurements are generally higher in the first condition)  
- $RBC < 0$  Negative differences tend to have larger ranks (e.g., measurements are generally higher in the second condition)  
- $|RBC| \approx 1$  Very strong directional tendency

**Why its useful:**  
RBC is **directly interpretable**:  
For example, $RBC = 0.156$ in our results means there is only a **small tendency** for the Before measurements to be higher than the After measurements, and this tendency is weak and likely due to chance (p-value = 0.397).


#### Common Language Effect Size (CLES)  

**Common Language Effect Size (CLES)** is an **effect size** measure that expresses results as an intuitive probability.

For the Wilcoxon signed-rank test with paired data, it answers:  
> *If we randomly select one non-zero paired difference, what is the probability that it is positive (i.e., the first condition has a higher value than the second)?*

**Properties:**  
- $CLES = 0.50$  Positive and negative differences are equally likely (no tendency)  
- $CLES > 0.50$  Positive differences are more likely (first condition tends to be higher)  
- $CLES < 0.50$  Negative differences are more likely (second condition tends to be higher)  
- Values near $1.0$ or $0.0$  Almost complete directional dominance

**Why its useful:**  
CLES converts the result into **plain language** that is easy to interpret without statistical training.  
For example, our $CLES = 0.535$ means there is a **53.5% probability** that a randomly chosen non-zero paired difference is positive  only a slight tendency for Before measurements to be higher than After, which is weak and likely due to chance (p-value = 0.397).


### Manual Calculation of Wilcoxon W-test

```python
from scipy.stats import rankdata, norm

# ----------------------------
# Wilcoxon Signed-Rank (paired)
# ----------------------------
# Paired differences (drop exact zeros)
diffs = (df["Before"] - df["After"]).astype(float)
d = diffs[diffs != 0].to_numpy()
n = len(d)

# Ranks of absolute differences (average ranks for ties)
r = rankdata(np.abs(d), method="average")

# Sum ranks for positive and negative differences
W_pos = r[d > 0].sum()
W_neg = r[d < 0].sum()
W = min(W_pos, W_neg)  # Wilcoxon W statistic

# Expected mean and std under H0 (no shift); tie correction optional
mean_W = n * (n + 1) / 4
std_W  = np.sqrt(n * (n + 1) * (2*n + 1) / 24)

# Z-score (continuity correction optional; here none)
z_score = (W - mean_W) / std_W

# Critical Z for two-tailed alpha=0.05
alpha = 0.05
critical_z = norm.ppf(1 - alpha/2)

# ----------------------------
# Plot Z distribution
# ----------------------------
plt.figure(figsize=(6,4))
xs = np.linspace(-4, 4, 1000)
plt.plot(xs, norm.pdf(xs), 'k', label="Z-Distribution PDF")
plt.axvline(critical_z, ymin=0.045, ymax=0.95, color='red', label='Critical Z-scores')
plt.axvline(-critical_z, ymin=0.045, ymax=0.95, color='red')
plt.axvline(z_score, ymin=0.045, ymax=0.95, color='blue', label=f'Observed Z = {z_score:.2f}')
plt.title("Wilcoxon Signed-Rank Test Critical Region")
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------------
# Print results
# ----------------------------
print(f"n (non-zero diffs) = {n}")
print(f"W_pos = {W_pos:.3f}, W_neg = {W_neg:.3f}")
print(f"W (min) = {W:.3f}")
print(f"Mean(W) = {mean_W:.3f}, Std(W) = {std_W:.3f}")
print(f"Z-score = {z_score:.3f}")
print(f"Critical Z-scores: {critical_z:.3f}")

# Optional: two-sided p-value from Z (normal approx)
p_val = 2 * (1 - norm.cdf(abs(z_score)))
print(f"Approx. two-sided p-value = {p_val:.6f}")
```

## Conclusion

A Wilcoxon signed-rank test was conducted to compare paired weight measurements taken **before** and **after** the intervention. Assumptions of paired measurements, symmetric distribution of differences, ordinal or higher scale, and absence of extreme outliers were all met.

Results showed **no statistically significant difference** in weights between the two conditions, $W = 346.0$, $z = -0.85$, $p = 0.397$. The median paired difference was **+1.0 lbs** (slightly higher after the intervention). Since the p-value is above the $5\%$ significance level, we fail to reject the null hypothesis and conclude there is no evidence of a systematic change in weight.

The effect size, rank-biserial correlation ($r_{rb} = 0.156$), indicates a very small tendency for weights to be higher before the intervention. The Common Language Effect Size ($CLES = 0.535$) suggests that in approximately **53.5%** of random non-zero difference pairs, the Before measurement is greater than the After measurement  a tendency likely due to chance.

**These results suggest that the intervention did not produce a meaningful change in participant weights.**
