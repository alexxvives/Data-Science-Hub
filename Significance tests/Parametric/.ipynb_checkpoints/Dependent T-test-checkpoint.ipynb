# Paired/Dependent T-test

A dependent (paired) t‑test is a **parametric** test that compares the **means of two related measurements** taken on the same units (or on matched pairs). *Related* means each value in condition A has a natural partner in condition B (e.g., the same person measured twice, or two carefully matched individuals).
 
It answers the question:
> *Is the average within‑pair difference equal to zero (or to a specified value)?*

Common scenarios include:
* The same individuals measured under two different conditions.
* The same individuals measured before and after an intervention. (SHOWN BELOW)
* Matched pairs of individuals in different groups, matched on variables such as age, gender, or background.

### Data Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats
import pingouin as pg
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
np.random.seed(42) # Seed for reproducibility
```

### Generate Data

```python
# Dataset of the change in weight (lbs) before and after completing a diet program
N = 40
before_data = np.random.normal(180, 15, N).astype(int)  # Before weight (lbs)
after_data = (before_data - np.random.normal(8, 4, N)).astype(int)  # After diet (lbs)
df = pd.DataFrame({"Pair": range(1, N+1), "Before": before_data, "After": after_data}).set_index("Pair")

# Take a look at the data
print(f"The dataset contains two weight measurements (before vs. after the diet) from {df.shape[0]} participants.")
display(HTML(
    f"""
    <div style="display: flex; gap: 50px;">
        <div>{df.head().to_html()}</div>
        <div>{df.describe().T.to_html()}</div>
    </div>
    """
))

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.kdeplot(ax=axes[0], x=df['Before'], shade=True, label='Before')
sns.kdeplot(ax=axes[0], x=df['After'], shade=True, label='After')
axes[0].set_title("KDE plot")
axes[0].set_xlabel('Systolic blood pressure')
axes[0].legend()
sns.boxplot(ax=axes[1], data=df)
axes[1].set_title('Boxplot per test')
plt.show()
```

### Hypothesis

$H_0$: The difference in weight before and after completing the diet program is equal to zero.<br>
$H_a$: The difference in weight before and after completing the diet program is not equal to zero.

**Assumptions**
* Assumption 1: The dependent variable is continuous (interval or ratio scale).
* Assumption 2: Observations are paired/matched and pairs are independent of each other.
* Assumption 3: The differences contain no extreme outliers (could disproportionately influence the mean)
* Assumption 4: The differences between paired observations are approximately normally distributed.

*Note: Of these, only the normality assumption can be mildly violated without greatly affecting results, particularly with sample sizes of 30 pairs or more; the other assumptions are essential for the validity of the paired t-test.*

```python
# Calculate the difference in weight before and after
diff = df['After'] - df['Before']
```

#### Assumption 1
This assumption is valid because weight is measured on a ratio scale in pounds, which is a continuous variable with equal intervals and a true zero.

#### Assumption 2
This assumption is valid because:<br>
**Matching pairs**: Each Before and After measurement belongs to the same individual, so they form a natural dapair.<br>
**Indepenent Pairs**: Each participant's results are unrelated to any other participant’s results — no repeated subjects, no hared environments that would artificially link pairs.

#### Assumption 3
This assumption is valid because no outliers were found. If any had been found, they should be investigated to ensure they are not data entry errors, and, if valid, the entire dataset could be transformed to reduce their influence on the test or the analysis could switch to a nonparametric alternative.

```python
# Calculate outliers 
Q1, Q3 = np.percentile(diff, [25, 75])
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = diff[(diff < lower_bound) | (diff > upper_bound)]
print(f"There are {len(outliers)} ouliers: {outliers.values}", )

# Plot boxplot (+ ouliers)
plt.figure(figsize=(3,4))
sns.boxplot(data=diff) 
sns.scatterplot(y=outliers, x=np.zeros(len(outliers)), color="red", s=60, label="Outliers")
plt.ylabel("Weight differences")
plt.title("Boxplot of weight differences")
plt.legend()
plt.show()
```

#### Assumption 4
This assumption is valid because it passed both a visual check using a QQ plot against normal (Gaussian) quantiles and a formal Shapiro–Wilk test.

```python
# Plot histogram and QQplot
fig, axes = plt.subplots(1, 2, figsize=(10, 3))
sns.histplot(diff, bins=10, ax=axes[0])
axes[0].set_xlabel("Difference (After - Before)")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Histogram of Differences")
sm.qqplot(diff, dist=stats.norm, line='s', ax=axes[1])
axes[1].set_title("QQ Plot of Differences")
plt.show()
```

A QQ plot (Quantile–Quantile plot) is a **visual** way to check whether your data follows a certain theoretical distribution, in this case, the normal distribution.<br> The red line represents where the points would fall if the data’s distribution matched the theoretical distribution perfectly.

```python
# Shapiro–Wilk test
shapiro_stat, shapiro_p = stats.shapiro(diff)
print(f"Shapiro-Wilk Test: W={shapiro_stat:.3f}, p={shapiro_p:.3f}")
```

SHAPIRO–WILK test<br>
$p > 0.05 \; \Rightarrow$ Fail to reject normality (assumption is plausible).  
$p < 0.05 \; \Rightarrow$ Evidence of non-normality; consider transformations or a nonparametric test (e.g., Wilcoxon signed-rank).

Therefore, we **fail to reject normality** (i.e. data is most likely normal).


## Dependent T-test

```python
result = pg.ttest(df['Before'], df['After'], paired=True)
result
```

#### P-value
If the null hypothesis is true (the diet has no effect), the probability of observing a mean weight change as extreme as this purely by chance is approximately 1.68e-14, which is far smaller than the conventional 0.05 significance level. Since the p-value is well below 0.05, **we reject the null hypothesis** and conclude that the diet program likely had a real effect on participants’ weight.

#### T statistic

The T statistic is the **ratio between the mean difference** (between the two measurements of the same subjects) and the **standard error of that difference**.<br>
Larger T $\Rightarrow$ the observed mean difference is large relative to the variation in the differences.<br>
Smaller T $\Rightarrow$ the observed mean difference is small compared to the variation, suggesting more similarity.

$$
T = \frac{\overline{d} - \mu}{SE_d}
  = \frac{\overline{d} - \mu}{\frac{s_d}{\sqrt{n}}}
$$

* *Where $\overline{d}$ is the mean of the difference* <br>
* $\mu$ is the hypothesized population mean difference ($0$ in this case)
* *$s_d$ is the standard deviation of the difference* <br>
* *n is the sample size* <br>

**Clear explanation:**<br>
Our null hypothesis is that the diet has no effect (i.e. the participants did not lose a statistically significant amount of weight) and therefore the mean of the weight differences (After-Before) is 0. The **numerator** of this formula represents the difference between the actual difference ($d$) and the expected mean difference ($\mu$, which is $0$). Therefore if the numerator is large (larger T) it means there's a notable average change between Before and After.<br>

But of course, there is always the risk that this result is the result of random chance, so the **denominator** represents how much the sample mean difference is expected to vary just by random chance, based on the standard deviation of the differences ($s_d$) and the number of pairs ($n$). The smaller the sample size, the likelier it is that the samples were just a few lucky pairs which in turn increases the Standard Error (see graph below), minimizing the effect of the difference in means (the numerator). 

```python
# Visual to show how SE decreases as sample size increases
sd_d = 12 # Assume SD of differences is 12 units for illustration
n_values = np.arange(5, 101, 5) # Sample sizes from 5 to 100 in steps of 5

# Calculate SE for each n
se_values = sd_d / np.sqrt(n_values)

# Plot
plt.figure(figsize=(5,2))
plt.plot(n_values, se_values, marker='o')
plt.xlabel("Number of pairs (n)")
plt.ylabel("Standard Error (SE)")
plt.title("Standard Error decreases as sample size increases")
plt.grid(True, alpha=0.3)
plt.show()
```

#### Confidence Interval

If we took 100 random samples from the population, we would expect about 95 of them to have an average weight loss between 6.97 lbs and 9.83 lbs.

#### Cohen's D

The test's Cohen d was 0.53, indicating a **medium effect** size according to Cohen's conventions. This suggests that the weight loss produced by the diet program is meaningful in magnitude, not just statistically significant.

While the p-value only tells you whether an effect is statistically significant or not, Cohen's D tells us how big the effect is (d=0.2 is considered a 'small' effect size, 0.5 is a 'medium' effect size, and that 0.8 is a 'large' effect size). <br>This means that if two groups' means don't differ by 0.2 standard deviations or more, the difference is trivial, even if it is statistically significant.

#### Power

Based on the observed effect size (Cohen's $d = 0.53$), sample size ($n = 40$), and significance level ($\alpha = 0.05$), the paired t-test had an estimated statistical power above $90\%$. This means the study had a very high probability of detecting a true weight-loss effect of this magnitude, reducing the risk of a Type II error.

The power answers the question: "Given there is a real effect, how likely am I to detect it?" (The probability of correctly rejecting $𝐻_0$ when $𝐻_a$ is actually true.)

### Manual Calculation of Dependent T-Test

```python
# Add variables  
SE = diff.std() / np.sqrt(len(df)) # Standard Error
t_score = (diff.mean() - 0) / SE
dof = df.shape[0]-1 # Degrees of freedom

# Significance level and tails
alpha = 0.05
ntails = 2

## Calculate critical t
critical_t = stats.t.ppf(q=1-(alpha/2),df=dof)

# Plot the t distribution
plt.figure()
xs = np.linspace(-15,15,1000)
plt.plot(xs, stats.t.pdf(xs,dof), 'k', label="T-Distribution PDF")

# Critical t cutoffs
plt.axvline(critical_t, ymin=0.045, ymax=0.95, color='red', label='Critical T-scores')
if ntails == 2: 
    plt.axvline(-critical_t, ymin=0.045, ymax=0.95, color='red')
plt.axvline(t_score, ymin=0.045, ymax=0.95, color='blue', label='T-score')
plt.legend()
plt.show()

print('Critical T-score: +-{} and T-score: {}'.format(round(critical_t,3),round(float(t_score),3)))
```

### Conclusion

The dependent t-test was conducted to compare participants' weight before and after completing the diet program, with differences defined as After - Before. Assumptions of continuity, paired independence, approximate normality of differences, and absence of extreme outliers were all met.

Results showed a statistically significant reduction in weight, $t(39)=11.85$, $p < 0.001$, with a mean weight loss of $M = 8.40$ lbs and a 95\% confidence interval of [6.97, 9.83] lbs. Since the entire CI lies above zero, we can be confident that the true average weight loss in the population is positive.

The effect size, Cohen's $d = 0.53$, indicates a medium magnitude of weight loss. The estimated statistical power ($>90\%$) suggests the study had a very high probability of detecting an effect of this size, minimizing the risk of a Type II error.

**These results provide strong evidence that the diet program produces a meaningful and reliable reduction in participants' weight.**
