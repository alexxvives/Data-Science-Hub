# Independent T-test

An independent (two-sample) t-test is a **parametric** test that compares the **means of two independent groups**. <br>
*Independent* means the observations in one group have no natural pairing or relationship with those in the other group.

It answers the question:
> *Is the difference between the two group means equal to zero (or to a specified value)?*

Common scenarios include:
* Comparing measurements from two different groups (e.g., treatment vs. control). (SHOWN BELOW)
* Comparing performance of two independent samples, such as students from two different schools.

## Data Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats
import pingouin as pg
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
np.random.seed(42) # Seed for reproducibility
```

```python
# Dataset of exam scores (%) for students after completing either interactive lessons or traditional lectures
N = 20
control = np.random.normal(74, 8, N).astype(int) # Traditional lectures
treatment = np.random.normal(80, 8, N).astype(int) # Interactive lessons
df = pd.DataFrame({"Student": range(1, N+1), "Interactive": treatment, "Traditional": control}).set_index("Student")

# Take a look at the data
print(f"The dataset contains exam scores from two independent groups: Interactive education (n={N}) and Traditional education (n={N}).")
display(HTML(
    f"""
    <div style="display: flex; gap: 50px;">
        <div>{df.head().to_html()}</div>
        <div>{df.describe().T.to_html()}</div>
    </div>
    """
))

# Visualizations (no tests)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# KDE plot
sns.kdeplot(ax=axes[0], x=df['Interactive'], fill=True, label='Interactive')
sns.kdeplot(ax=axes[0], x=df['Traditional'], fill=True, label='Traditional')
axes[0].set_title("KDE plot")
axes[0].set_xlabel('Exam Score')
axes[0].legend()

# Boxplot
sns.boxplot(ax=axes[1], data=df)
axes[1].set_title('Boxplot per group')
axes[1].set_ylabel('Exam Score')

plt.show()
```

### Hypothesis

$H_0$: The mean exam score for students who completed interactive lessons is equal to the mean score for students who completed traditional lectures.<br>
$H_a$: The mean exam scores for the two groups are not equal.

**Assumptions**
* Assumption 1: The dependent variable is continuous (interval or ratio scale).
* Assumption 2: Observations are independent between and within groups (no participant appears in both groups).
* Assumption 3: The data contains no extreme outliers (could disproportionately influence the mean)
* Assumption 4: The dependent variable is normally distributed within each group.

#### Assumption 1
This assumption is valid because exam scores are measured on a ratio scale from 0 to 100, which is a continuous variable with equal intervals and a true zero.

#### Assumption 2
This assumption is valid because:<br>
**Independence between groups:** Each student belongs to only one group, either Interactive or Traditional, with no overlap.<br>
**Independence within groups:** Each students exam score is unrelated to the scores of other students in the same group

#### Assumption 3
This assumption is valid because no outliers were found. If any had been found, they should be investigated to ensure they are not data entry errors, and, if valid, the entire dataset could be transformed to reduce their influence on the test or the analysis could switch to a nonparametric alternative.

```python
# Reshape to long format for plotting
df_long = df.melt(var_name="Group", value_name="Score")

plt.figure(figsize=(5,5))
palette = ["#3D75A6", "#A63D3D"]
legend_labels = ["Interactive (Treatment)", "Traditional (Control)"]
sns.boxplot(data=df_long, x="Group", y="Score", palette=palette, flierprops={"markerfacecolor": "red", "marker": "o", "markersize": 6})
patches = [mpatches.Patch(color=palette[i], label=legend_labels[i]) for i in range(len(palette))]
plt.title("Boxplot of Exam Scores with Outliers Highlighted")
plt.ylabel("Exam Score")
plt.legend(handles=patches, title="Group", loc="upper right")

plt.show()
```

#### Assumption 4
This assumption is valid because it passed both a visual check using a QQ plot against normal (Gaussian) quantiles and a formal ShapiroWilk test.

```python
for group in df.columns:
    scores = df[group]
    
    # Plot histogram and QQ plot
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    sns.histplot(scores, bins=8, ax=axes[0])
    axes[0].set_xlabel("Exam Score")
    axes[0].set_ylabel("Frequency")
    axes[0].set_title(f"Histogram of {group} Scores")
    
    sm.qqplot(scores, dist=stats.norm, line='s', ax=axes[1])
    axes[1].set_title(f"QQ Plot of {group} Scores")
    plt.show()
```

A QQ plot (QuantileQuantile plot) is a **visual** way to check whether your data follows a certain theoretical distribution, in this case, the normal distribution.<br> The red line represents where the points would fall if the datas distribution matched the theoretical distribution perfectly.

```python
# ShapiroWilk test
for group in df.columns:
    scores = df[group]
    
    shapiro_stat, shapiro_p = stats.shapiro(scores)
    print(f"{group} - Shapiro-Wilk Test: W={shapiro_stat:.3f}, p={shapiro_p:.3f}")
```

SHAPIROWILK test<br>
$p > 0.05 \; \Rightarrow$ Fail to reject normality (assumption is plausible).  
$p < 0.05 \; \Rightarrow$ Evidence of non-normality; consider transformations or a nonparametric test (e.g., MannWhitney U test for independent samples).

Therefore, we **fail to reject normality** for both groups (i.e. data is most likely normal).


## Independent T-test

```python
# Run Welchs t-test (correction=True), which doesn't assume equal population variances (opposed to pooled independent t-test)
result = pg.ttest(df['Interactive'], df['Traditional'], paired=False, correction=True)
result
```

#### P-value
If the null hypothesis is true (there is no difference in mean exam scores between the Interactive and Traditional groups), the probability of observing a difference in means as extreme as the one found purely by chance is approximately 3.9%. Since this p-value is below the conventional 5% significance level, we **reject the null hypothesis** and conclude that the alternative teaching method likely had a real effect on students' exam performance.

#### T statistic

The T statistic is the **ratio between the difference in sample means** (between the two independent groups) and the **standard error of that difference**.  
Larger $T \Rightarrow$ the observed mean difference is large relative to the variation in the data.  
Smaller $T \Rightarrow$ the observed mean difference is small compared to the variation, suggesting more similarity between groups.  

$$
T = \frac{\overline{X}_1 - \overline{X}_2}{SE_{\text{Welch}}}
$$

For **Welchs t-test**, the standard error is calculated as:

$$
SE_{\text{Welch}} = \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}
$$

* $\overline{X}_1$ and $\overline{X}_2$ are the sample means of the two groups  
* $s_1^2$ and $s_2^2$ are the sample variances of the two groups  
* $n_1$ and $n_2$ are the sample sizes of each group  

**Clear explanation:**  
Our null hypothesis is that the teaching method has no effect (i.e., the average exam scores are the same for the Interactive and Traditional groups). The **numerator** of this formula represents the observed difference between the two group means. If this difference is large relative to the expected variability, the $T$ value will be large.  

The **denominator** represents how much variation in the difference wed expect purely from random sampling, based on the variability within each group and their respective sample sizes. Smaller sample sizes or higher within-group variability increase the standard error, which reduces the $T$ statistic and makes it harder to detect a significant difference.


```python
# Visual to show how SE decreases as sample size increases
sd1, sd2 = 10, 12 
n_values = np.arange(5, 101, 5)

# Calculate SE for Welch's t-test for each n
se_values = np.sqrt((sd1**2 / n_values) + (sd2**2 / n_values))

# Plot
plt.figure(figsize=(5, 2.5))
plt.plot(n_values, se_values, marker='o')
plt.xlabel("Sample size per group (n)")
plt.ylabel("Standard Error (SE)")
plt.title("SE (Welch) decreases as sample size increases")
plt.grid(True, alpha=0.3)
plt.show()
```

#### Confidence Interval
If we took 100 random samples from the population, we would expect about 95 of them to have a mean difference in exam scores between 0.32 and 12.48 points.

#### Cohen's D

The test's Cohen d was 0.67, indicating a **medium effect** size according to Cohen's conventions. This suggests that the score increase on the treatment group is meaningful in magnitude, not just statistically significant.

While the p-value only tells you whether an effect is statistically significant or not, Cohen's D tells us how big the effect is (d=0.2 is considered a 'small' effect size, 0.5 is a 'medium' effect size, and that 0.8 is a 'large' effect size). <br>This means that if two groups' means don't differ by 0.2 standard deviations or more, the difference is trivial, even if it is statistically significant.

#### Power

Based on the observed effect size (Cohen's $d = 0.68$), sample size per group ($n = 20$), and significance level ($\alpha = 0.05$), the Welch's independent t-test had an estimated statistical power of approximately $54\%$. This means the study had a high probability of detecting a true difference in exam scores of this magnitude, reducing the risk of a Type II error.

The power answers the question: "Given there is a real difference between the groups, how likely am I to detect it?"  in other words, the probability of correctly rejecting $H_0$ when $H_a$ is actually true.

### Manual Calculation of Independent T-Test

```python
# Add variables 
group1 = df['Interactive']
group2 = df['Traditional']
SE = np.sqrt(group1.var(ddof=1)/len(group1) + group2.var(ddof=1)/len(group2)) # Welch's standard error
t_score = (group1.mean() - group2.mean()) / SE # T-score for Welch's test
s1, s2 = group1.var(ddof=1), group2.var(ddof=1)
n1, n2 = len(group1), len(group2)
dof = (s1/n1 + s2/n2)**2 / ((s1**2) / ((n1**2)*(n1-1)) + (s2**2) / ((n2**2)*(n2-1)))

# Significance level and tails
alpha = 0.05
ntails = 2

# Critical t value
critical_t = stats.t.ppf(q=1 - (alpha/2), df=dof)

# Plot the t distribution
plt.figure(figsize=(6,4))
xs = np.linspace(-5, 5, 1000)
plt.plot(xs, stats.t.pdf(xs, dof), 'k', label="T-Distribution PDF")

# Critical t cutoffs
plt.axvline(critical_t, ymin=0.045, ymax=0.95, color='red', label='Critical T-scores')
if ntails == 2:
    plt.axvline(-critical_t, ymin=0.045, ymax=0.95, color='red')
plt.axvline(t_score, ymin=0.045, ymax=0.95, color='blue', label='Observed T-score')
plt.legend()
plt.title("Welch's Independent T-Test Critical Region")
plt.show()

print(f'Critical T-score: +-{critical_t:.3f} and T-score: {t_score:.3f}')
```

## Conclusion

A Welch's independent samples t-test was conducted to compare exam scores between students who completed interactive lessons (treatment) and those who completed traditional lectures (control). Assumptions of continuity, independence between groups, approximate normality within each group, and absence of extreme outliers were all met.  

Results showed a statistically significant difference in mean exam scores, $t(37.996) = 2.14$, $p = 0.039$, with the interactive lessons group scoring on average $M_{\text{diff}} = 5.20$ points higher than the traditional lectures group. The 95\% confidence interval for the mean difference was $[0.27, 10.13]$ points, and since the entire CI lies above zero, we can be confident that the true mean difference in the population is positive.  

The effect size, Cohen's $d = 0.68$, indicates a medium-to-large difference in performance between teaching methods. The estimated statistical power ($\approx 54\%$) indicates a modest probability of detecting an effect of this size, meaning there remains a considerable risk of a Type II error.  

**These results provide evidence that interactive lessons improve exam performance compared to traditional lectures, though the relatively low power suggests the finding should be interpreted with caution and replicated with a larger sample.**
