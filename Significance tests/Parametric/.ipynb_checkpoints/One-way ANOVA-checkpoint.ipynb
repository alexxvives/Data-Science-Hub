# One-way ANOVA

A one-way ANOVA (Analysis of Variance) is a **parametric test** used to compare the means of **three or more independent groups** to determine whether at least one group mean is significantly different from the others.

It answers the question: *Is there evidence that not all group means are equal?*

Common scenarios include:
* Comparing test scores across three different teaching methods.
* Evaluating production yields from multiple factories.
* Comparing customer satisfaction scores among several service providers. (SHOWN BELOW)

## Data Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import numpy.random as rnd
from random import sample
import statsmodels.api as sm
import scipy.stats as stats
import pingouin as pg
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")

sns.set_style('darkgrid')
rng = np.random.default_rng(42)
```

```python
# Sample size per group
N = 20

# Simulated people heights
np.random.seed(42)  # Seed for reproducibility
country_A = np.random.normal(175, 5, N).astype(int)
country_B = np.random.normal(180, 4, N).astype(int)
country_C = np.random.normal(165, 5, N).astype(int)

# Create dataframe
df = pd.DataFrame({
    "ID": range(1, N+1),
    "Country A": country_A,
    "Country B": country_B,
    "Country C": country_C
}).set_index("ID")

# Overview of dataset
print(f"The dataset contains people's heights for three different countries: "
      f"country A (n={N}), country B (n={N}), and country C (n={N}).")

display(HTML(
    f"""
    <div style="display: flex; gap: 50px;">
        <div>{df.head().to_html()}</div>
        <div>{df.describe().T.to_html()}</div>
    </div>
    """
))

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# KDE plots for each provider
for country in df.columns:
    sns.kdeplot(ax=axes[0], x=df[country], fill=True, label=country)
axes[0].set_title("KDE Plot of Heights Scores by Country")
axes[0].set_xlabel("Height (cm)")
axes[0].legend()

# Boxplot
sns.boxplot(ax=axes[1], data=df)
axes[1].set_title("Boxplot of Heights by Country")
axes[1].set_ylabel("Height (cm)")

plt.show()
```

### Hypothesis

$H_0$: The mean heights across all three countries are equal.  
$H_a$: At least one country's mean height is different from the others.

**Assumptions**
* Assumption 1: The dependent variable is continuous (interval or ratio scale).
* Assumption 2: Observations are independent
* Assumption 3: The data contains no extreme outliers (could disproportionately influence the mean)
* Assumption 4: The satisfaction scores for each provider are approximately normally distributed.
* Assumption 5: The variability of satisfaction scores is roughly equal across all provider groups (Homogeneity of variances).

#### Assumption 1
This assumption is valid because height is measured on a ratio scale in centimeters (cm), which is a continuous variable with equal intervals and a true zero.

#### Assumption 2  
This assumption is valid because:  
**Independence between groups:** Each participants height was recorded for only one country group, with no individual appearing in more than one group.  
**Independence within groups:** Each participants height measurement is independent of the measurements from other participants in the same country.  

#### Assumption 3
This assumption is valid because no outliers were found. If any had been found, they should be investigated to ensure they are not data entry errors, and, if valid, the entire dataset could be transformed to reduce their influence on the test or the analysis could switch to a nonparametric alternative.

```python
# Reshape to long format for plotting
df_long = df.melt(var_name="Country", value_name="Height")

# Boxplot
plt.figure(figsize=(6,5))
palette = ["#3D75A6", "#A63D3D", "#3DA66E"]
legend_labels = ["Country A", "Country B", "Country C"]
sns.boxplot(data=df_long, x="Country", y="Height", palette=palette,flierprops={"markerfacecolor": "red", "marker": "o", "markersize": 6})
patches = [mpatches.Patch(color=palette[i], label=legend_labels[i]) for i in range(len(palette))]
plt.title("People's heights by Country (with Outliers Highlighted)")
plt.ylabel("Height (cm)")
plt.legend(handles=patches, title="Country", loc="upper right")
plt.show()
```

#### Assumption 4
This assumption is valid because it passed both a visual check using a QQ plot against normal (Gaussian) quantiles and a formal ShapiroWilk test.

```python
for country in df.columns:
    heights = df[country]

    # Plot histogram and QQ plot
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    sns.histplot(heights, bins=10, ax=axes[0])
    axes[0].set_xlabel("Height (cm)")
    axes[0].set_ylabel("Frequency")
    axes[0].set_title(f"Histogram of {country} Heights")

    sm.qqplot(heights, dist=stats.norm, line='s', ax=axes[1])
    axes[1].set_title(f"QQ Plot of {country} Heights")
    plt.show()
```

A QQ plot (QuantileQuantile plot) is a **visual** way to check whether your data follows a certain theoretical distribution, in this case, the normal distribution.<br> The red line represents where the points would fall if the datas distribution matched the theoretical distribution perfectly.

```python
for country in df.columns:
    heights = df[country]
    
    # ShapiroWilk test
    shapiro_stat, shapiro_p = stats.shapiro(heights)
    print(f"{country} - Shapiro-Wilk Test: W={shapiro_stat:.3f}, p={shapiro_p:.3f}")
```

SHAPIROWILK test<br>
$p > 0.05 \; \Rightarrow$ Fail to reject normality (assumption is plausible).  
$p < 0.05 \; \Rightarrow$ Evidence of non-normality; consider transformations or a nonparametric test.

Therefore, we **fail to reject normality** for all groups (i.e. data is most likely normal).


#### Assumption 5

This assumption is valid because it passed a Bartletts test, indicating no statistically significant difference in variances across the three groups.

```python
# Bartlett's test
stat_bartlett, p_bartlett = stats.bartlett(df['Country A'], df['Country B'], df['Country C'])
print(f"Bartletts test: stat={stat_bartlett:.4f}, p={p_bartlett:.4f}")
```

## One-way ANOVA

```python
# One-way ANOVA
pg.anova(dv='Height', between='Country', data=df_long, detailed=True)
```

#### P-value  

If the null hypothesis is true (all countries have **the same mean height**), the probability of observing an $F$-statistic as extreme as the one obtained $(F(2,57) \approx 56.35)$ purely by chance is approximately $(p = 3.14 \times 10^{-14})$.

Since this p-value is far below the conventional (0.05) significance level, **we reject the null hypothesis** and conclude that **at least one country's mean height differs** from the others.


#### F-statistic

The F statistic is the **ratio between the variance explained by group differences** and the **variance within the groups**:<br>
Larger T $\Rightarrow$ the observed mean difference is large relative to the variation in the differences.<br>
Smaller T $\Rightarrow$ the observed mean difference is small compared to the variation, suggesting more similarity.

$$
F \;=\; \frac{MS_{\text{between}}}{MS_{\text{within}}}
\;=\; \frac{\frac{SS_{\text{between}}}{df_{\text{between}}}}{\frac{SS_{\text{within}}}{df_{\text{within}}}}
\;=\; \frac{\dfrac{\sum_{j=1}^k n_j \left(\overline{X}_j - \overline{X}_{\text{grand}}\right)^2}{k-1}}
         {\dfrac{\sum_{j=1}^k \sum_{i=1}^{n_j} \left(X_{ij} - \overline{X}_j\right)^2}{N-k}}
$$

Where:

- **$SS_{\text{between}}$** (Sum of Squares Between) measures the variability of each group mean from the overall mean.
- **$df_{\text{between}} = k - 1$** where $k$ is the number of groups.
- **$SS_{\text{within}}$** (Sum of Squares Within) measures the variability of the data points from their respective group means.
- **$df_{\text{within}} = N - k$** where $N$ is the total number of observations.

**Interpretation:**  
The numerator represents the difference between each group's means and the overall mean, and the denominator represents the spread of data points within each group. **If the numerator is low** (all groups's means are close to the overall mean), you would only reject the null hypothesis (ie. conclude that there is at least 1 coiuntry with diffferent mean height) if the variance within each group is very small. On the other hand, **if the numerator is high** you would likely reject the null hypothesis, unless there is a very large variance within groups.

#### Partial Eta Squared ($\eta_p^2$)

Partial eta squared is a **measure of effect size** used in ANOVA that represents the proportion of variance in the heights (dependent variable) that is **attributable to** the factor Country.
In a one-way ANOVA with only one factor, partial eta squared is equivalent to eta squared and reflects the total proportion of variance explained by the factor.

It answers:*What proportion of the variability in the dependent variable can be explained by differences between countries?*

$$
\eta_p^2 = \frac{SS_{\text{Between}}}{SS_{\text{Between}} + SS_{\text{Within}}}
$$
Where:
- $SS_{\text{Between}}$ = Sum of squares for the factor of interest  
- $SS_{\text{Within}}$ = Sum of squares for the error (within-group variation)  

**Interpretation guidelines** (Cohen, 1988):
- Small effect: $\eta_p^2 \approx 0.01$  
- Medium effect: $\eta_p^2 \approx 0.06$  
- Large effect: $\eta_p^2 \approx 0.14$


### Manual Calculation of a One-way ANOVA

```python
groups = [df['Country A'], df['Country B'], df['Country C']]
k = len(groups) 
ns = np.array([len(df['Country A']), len(df['Country B']), len(df['Country C'])])
N = ns.sum()
means = np.array([df['Country A'].mean(), df['Country B'].mean(), df['Country C'].mean()])
grand_mean = np.concatenate(groups).mean()

# Between-groups SS: sum over groups 
SS_between = np.sum(ns * (means - grand_mean)**2)

# Within-groups SS: sum over groups of sum_i (x_ij - mean_j)^2
SS_within = np.sum([np.sum((g - m)**2) for g, m in zip(groups, means)])

# Degrees of freedom
df_between = k - 1
df_within = N - k

# Mean Squares
MS_between = SS_between / df_between
MS_within  = SS_within  / df_within

# F statistic and p-value (upper-tail test)
F_obs = MS_between / MS_within
p_value = 1 - stats.f.cdf(F_obs, df_between, df_within)  # same as stats.f.sf(F_obs, ...)

# Critical F at alpha
alpha = 0.05
F_crit = stats.f.ppf(1 - alpha, df_between, df_within)

# ----------------------------
# Plot: F distribution with critical cutoff and observed F
# ----------------------------
plt.figure(figsize=(6,4))
x_max = max(F_crit, F_obs) * 1.5
xs = np.linspace(0, x_max, 1000)
plt.plot(xs, stats.f.pdf(xs, df_between, df_within), 'k', label='F-Distribution PDF')

# Critical line (rejection region to the right)
plt.axvline(F_crit, ymin=0.045, ymax=0.95, color='red', label=f'Critical F @ ={alpha:.2f}')
# Observed F
plt.axvline(F_obs, ymin=0.045, ymax=0.95, color='blue', label=f'Observed F = {F_obs:.2f}')

plt.title('One-way ANOVA Critical Region')
plt.xlabel('F value')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Groups (k): {k}, Total N: {N}")
print(f"SS_between = {SS_between:.3f}, MS_between = {MS_between:.3f}")
print(f"SS_within  = {SS_within:.3f}, MS_within  = {MS_within:.3f}")
print(f"F({df_between}, {df_within}) = {F_obs:.3f},  p = {p_value:.4f},  F_crit@={alpha:.2f} = {F_crit:.3f}")
```

### Conclusion

A one-way ANOVA was conducted to compare the mean values of the dependent variable across three groups. Assumptions of continuity, independence of observations, approximate normality within groups, and homogeneity of variances were considered to be met.  

Results showed a **statistically significant difference** between group means, $F(2, 57) = 56.35$, $p < 0.001$. The effect size, partial $\eta^2 = 0.664$, indicates that approximately **66.4\%** of the variance in the dependent variable is explained by group membership, representing a **very large** effect according to Cohens (1988) guidelines.  

**These results provide strong evidence that mean values differ significantly across the groups, with the factor Country accounting for the majority of variability in the dependent variable.**
